# I. def

> ## 1) 사용자 정의 함수

* hap( )

```
def hap(x, y):
    return x + y
```

```
hap(3, 6)
```

* List

```
L1 = [1, 3, hap, 9]

L1[2](5, 7)
```

* app( )

```
def app(func, x, y):
    return func(x, y)
```

```
app(hap, 2, 9)
```

* cha( ) with lambda

```
cha = lambda x, y : x - y

cha(3, 6)
```

# II. if else

```
n = 3
```

> ## 1) if

```
if n == 3:
    print('n은 3')
```

> ## 2) if else

```
if n != 3:
    print('Bye')
else:
    print('n은 3')
```

> ## 3) if elif else

```
if n < 2:
    print('A')
elif n < 5:
    print('B')
else:
    print('C')
```

# III. for in

> ## 1) for in with range( )

```
for i in range(1, 10):
    print(i)
```

> ## 2) for in with List_1

```
L1 = ['A', 'B', 'C', 'D', 'E']

for i in L1:
    print(i)
```

> ## 3) for in with List_2

```
L2 = [['A', 11], ['B', 22], ['C', 33]]

for i, j in L2:
    print(i)
    print(j)
```

> ## 4) for in with break

```
for i in range(1, 10):
    if i > 5: break
    print(i)
```

# V. while

```
i = 1
```

> ## 1) while else

```
while i < 10:
    print(i)
    i += 1
else:
    print('Fin')
```

```
i = 1
```

> ## 2) while if break

```
while i < 10:
    print(i)
    if i == 5:
        break
    i += 1
```
